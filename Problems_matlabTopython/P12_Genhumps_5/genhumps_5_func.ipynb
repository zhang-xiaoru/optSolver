```python
import numpy as np

# IOE 511/MATH 562, University of Michigan
# Code written by: Albert S. Berahas

# Problem Number: 12
# Problem Name: genhumps_5
# Problem Description: A multi-dimensional problem with a lot of humps.
#                      This problem is from the well-known CUTEr test set.

# function that computes the function value of the genhumps_5 function
def genhumps_5_func(x):

    f = 0
    
    # Ensure x is a 1D array to mimic MATLAB's linear indexing behavior
    x = np.array(x).flatten()

    # MATLAB: for i = 1:4
    # Python: range(4) generates 0, 1, 2, 3. 
    # This correctly maps x(1)..x(4) to x[0]..x[3]
    for i in range(4):
        # f = f + sin(2*x(i))^2*sin(2*x(i+1))^2 + 0.05*(x(i)^2 + x(i+1)^2);
        term1 = np.sin(2*x[i])**2 * np.sin(2*x[i+1])**2
        term2 = 0.05 * (x[i]**2 + x[i+1]**2)
        f = f + term1 + term2

    return f
```
