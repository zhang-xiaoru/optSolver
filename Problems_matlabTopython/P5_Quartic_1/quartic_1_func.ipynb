```python
import numpy as np

# IOE 511/MATH 562, University of Michigan
# Code written by: Albert S. Berahas

# Problem Number: 5
# Problem Name: quartic_1
# Problem Description: A quartic function. Dimension n = 4

# function that computes the function value of the quartic_1 function
def quartic_1_func(x):

    # Matrix Q
    # Q = [5 1 0 0.5;
    #      1 4 0.5 0;
    #      0 0.5 3 0;
    #      0.5 0 0 2];
    Q = np.array([
        [5, 1, 0, 0.5],
        [1, 4, 0.5, 0],
        [0, 0.5, 3, 0],
        [0.5, 0, 0, 2]
    ])
     
    # Set sigma value
    sigma = 1e-4

    # compute function value
    # f = 1/2*(x'*x) + sigma/4*(x'*Q*x)^2;
    
    # Ensure x is column vector for matrix operations
    x = np.array(x).reshape(-1, 1)
    
    # Calculate f
    f = 0.5 * (x.T @ x) + (sigma / 4) * (x.T @ Q @ x)**2

    return f.item()
```
